# Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
df.isnull().sum() 
```

```python
df.dropna(axis=0, how='all', inplace=True) 
```

```python
# REMOVE NULL VALUES OF PERTICULAR COLUMN 
df.fillna('Cylinders',inplace=True)    
```

```python
# Fill na value with "AB"
df.fillna("AB",inplace=True)   
df 
```

```python
# Replace nan value with backward value of column 
df.fillna(method="bfill",axis=1) 
```

## Find column names which have null values  

```python
a1= [i for i in df.columns if df[i].isnull().sum()>0]

# Convert list to string
a=" ".join(a1)    
```

```python
# replace with mean value of the column 
df[a1].fillna(df[a1].mean(), inplace=True)  
```


## Interpolate 

```python
# Replace missing value With predicted linear regression value of the column ====== change only numerical values  
df.interpolate(method='linear',inplace=True) 
```

```python
df.interpolate(method='time',inplace=True)
```

```python
df.isnull().sum() 
```

```python
df[a].interpolate(method = 'linear', inplace=True)  
```
